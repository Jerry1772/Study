# <span style="color:Salmon">CountFactors</span>

<div class="brinza-task-description">
<p>A positive integer D is a <i>factor</i> of a positive integer N if there exists an integer M such that N = D * M.</p>
<p>For example, 6 is a factor of 24, because M = 4 satisfies the above condition (24 = 6 * 4).</p>
<p>Write a function:</p>
<blockquote><p style="font-family: monospace; font-size: 9pt; display: block; white-space: pre-wrap;"><tt>def solution(N)</tt></p></blockquote>
<p>that, given a positive integer N, returns the number of its factors.</p>
<p>For example, given N = 24, the function should return 8, because 24 has 8 factors, namely 1, 2, 3, 4, 6, 8, 12, 24. There are no other factors of 24.</p>
<p>Write an <b><b>efficient</b></b> algorithm for the following assumptions:</p>
<blockquote><ul style="margin: 10px; padding: 0px;"><li>N is an integer within the range [<span class="number">1</span>..<span class="number">2,147,483,647</span>].</li>
</ul>
</blockquote></div>

````python
def solution(N):
    answer = 0
    root = N**0.5
    for i in range(1, int(root)+1):
        if N%i==0:
            answer += 2
    if root == int(root):
        answer -= 1
    return answer
````

# <span style="color:Salmon">MinPerimeterRectangle</span>

<div class="brinza-task-description">
<p>An integer N is given, representing the area of some rectangle.</p>
<p>The <i>area</i> of a rectangle whose sides are of length A and B is A * B, and the <i>perimeter</i> is 2 * (A + B).</p>
<p>The goal is to find the minimal perimeter of any rectangle whose area equals N. The sides of this rectangle should be only integers.</p>
<p>For example, given integer N = 30, rectangles of area 30 are:</p>
<blockquote><ul style="margin: 10px; padding: 0px;"><li>(1, 30), with a perimeter of 62,</li>
<li>(2, 15), with a perimeter of 34,</li>
<li>(3, 10), with a perimeter of 26,</li>
<li>(5, 6), with a perimeter of 22.</li>
</ul>
</blockquote><p>Write a function:</p>
<blockquote><p style="font-family: monospace; font-size: 9pt; display: block; white-space: pre-wrap;"><tt>def solution(N)</tt></p></blockquote>
<p>that, given an integer N, returns the minimal perimeter of any rectangle whose area is exactly equal to N.</p>
<p>For example, given an integer N = 30, the function should return 22, as explained above.</p>
<p>Write an <b><b>efficient</b></b> algorithm for the following assumptions:</p>
<blockquote><ul style="margin: 10px; padding: 0px;"><li>N is an integer within the range [<span class="number">1</span>..<span class="number">1,000,000,000</span>].</li>
</ul>
</blockquote></div>

````python
#     (y=-x+b  minimize b , under the condition y=a/x)
def solution(N):
    root = N**0.5
    for i in range(int(root), 0, -1):
        if N%i == 0:
            return 2*(i+N//i)
````

# <span style="color:Salmon">Flags</span>

<div class="brinza-task-description">
<p>A non-empty array A consisting of N integers is given.</p>
<p>A <i>peak</i> is an array element which is larger than its neighbours. More precisely, it is an index P such that 0 &lt; P &lt; N  1 and A[P  1] &lt; A[P] &gt; A[P + 1].</p>
<p>For example, the following array A:</p>
<tt style="white-space:pre-wrap">    A[0] = 1
    A[1] = 5
    A[2] = 3
    A[3] = 4
    A[4] = 3
    A[5] = 4
    A[6] = 1
    A[7] = 2
    A[8] = 3
    A[9] = 4
    A[10] = 6
    A[11] = 2</tt>
<p>has exactly four peaks: elements 1, 3, 5 and 10.</p>
<p>You are going on a trip to a range of mountains whose relative heights are represented by array A, as shown in a figure below. You have to choose how many flags you should take with you. The goal is to set the maximum number of flags on the peaks, according to certain rules.</p>
<p><img class="inline-description-image" style="background-color:white;" src="https://codility-frontend-prod.s3.amazonaws.com/media/task_static/flags/static/images/auto/6f5e8faa3000c0a74157e6e0bc759b8a.png"></p>
<p>Flags can only be set on peaks. What's more, if you take K flags, then the distance between any two flags should be greater than or equal to K. The distance between indices P and Q is the absolute value |P  Q|.</p>
<p>For example, given the mountain range represented by array A, above, with N = 12, if you take:</p>
<blockquote><ul style="margin: 10px;padding: 0px;"><li>two flags, you can set them on peaks 1 and 5;</li>
<li>three flags, you can set them on peaks 1, 5 and 10;</li>
<li>four flags, you can set only three flags, on peaks 1, 5 and 10.</li>
</ul>
</blockquote><p>You can therefore set a maximum of three flags in this case.</p>
<p>Write a function:</p>
<blockquote><p style="font-family: monospace; font-size: 9pt; display: block; white-space: pre-wrap"><tt>def solution(A)</tt></p></blockquote>
<p>that, given a non-empty array A of N integers, returns the maximum number of flags that can be set on the peaks of the array.</p>
<p>For example, the following array A:</p>
<tt style="white-space:pre-wrap">    A[0] = 1
    A[1] = 5
    A[2] = 3
    A[3] = 4
    A[4] = 3
    A[5] = 4
    A[6] = 1
    A[7] = 2
    A[8] = 3
    A[9] = 4
    A[10] = 6
    A[11] = 2</tt>
<p>the function should return 3, as explained above.</p>
<p>Write an <b><b>efficient</b></b> algorithm for the following assumptions:</p>
<blockquote><ul style="margin: 10px;padding: 0px;"><li>N is an integer within the range [<span class="number">1</span>..<span class="number">400,000</span>];</li>
<li>each element of array A is an integer within the range [<span class="number">0</span>..<span class="number">1,000,000,000</span>].</li>
</ul>
</blockquote></div>

````python
#   flag  ? -> min(root(n), peaks) ((root(n)   n))

def solution(A):
    peaks = []
    for idx in range(1, len(A)-1):
        if A[idx]>A[idx-1] and A[idx]>A[idx+1]:
            peaks.append(idx)

    iterations = min(len(peaks), int(len(A)**0.5)+1)
    
    for flags in range(iterations, 0, -1):
        last_peak = -1e+6
        remains = flags
        for peak in peaks:
            if peak-last_peak>=flags:
                remains -= 1
                last_peak = peak
            if not remains:
                return flags
    return 0
````

# <span style="color:Salmon">Peaks</span>

<div class="brinza-task-description">
<p>A non-empty array A consisting of N integers is given.</p>
<p>A <i>peak</i> is an array element which is larger than its neighbors. More precisely, it is an index P such that 0 &lt; P &lt; N  1,&nbsp; A[P  1] &lt; A[P] and A[P] &gt; A[P + 1].</p>
<p>For example, the following array A:</p>
<tt style="white-space: pre-wrap;">    A[0] = 1
    A[1] = 2
    A[2] = 3
    A[3] = 4
    A[4] = 3
    A[5] = 4
    A[6] = 1
    A[7] = 2
    A[8] = 3
    A[9] = 4
    A[10] = 6
    A[11] = 2</tt>
<p>has exactly three peaks: 3, 5, 10.</p>
<p>We want to divide this array into blocks containing the same number of elements. More precisely, we want to choose a number K that will yield the following blocks:</p>
<blockquote><ul style="margin: 10px; padding: 0px;"><li>A[0], A[1], ..., A[K  1],</li>
<li>A[K], A[K + 1], ..., A[2K  1],<br>
 ...</li>
<li>A[N  K], A[N  K + 1], ..., A[N  1].</li>
</ul>
</blockquote><p>What's more, every block should contain at least one peak. Notice that extreme elements of the blocks (for example A[K  1] or A[K]) can also be peaks, but only if they have both neighbors (including one in an adjacent blocks).</p>
<p>The goal is to find the maximum number of blocks into which the array A can be divided.</p>
<p>Array A can be divided into blocks as follows:</p>
<blockquote><ul style="margin: 10px; padding: 0px;"><li>one block (1, 2, 3, 4, 3, 4, 1, 2, 3, 4, 6, 2). This block contains three peaks.</li>
<li>two blocks (1, 2, 3, 4, 3, 4) and (1, 2, 3, 4, 6, 2). Every block has a peak.</li>
<li>three blocks (1, 2, 3, 4), (3, 4, 1, 2), (3, 4, 6, 2). Every block has a peak. Notice in particular that the first block (1, 2, 3, 4) has a peak at A[3], because A[2] &lt; A[3] &gt; A[4], even though A[4] is in the adjacent block.</li>
</ul>
</blockquote><p>However, array A cannot be divided into four blocks, (1, 2, 3), (4, 3, 4), (1, 2, 3) and (4, 6, 2), because the (1, 2, 3) blocks do not contain a peak. Notice in particular that the (4, 3, 4) block contains two peaks: A[3] and A[5].</p>
<p>The maximum number of blocks that array A can be divided into is three.</p>
<p>Write a function:</p>
<blockquote><p style="font-family: monospace; font-size: 9pt; display: block; white-space: pre-wrap;"><tt>def solution(A)</tt></p></blockquote>
<p>that, given a non-empty array A consisting of N integers, returns the maximum number of blocks into which A can be divided.</p>
<p>If A cannot be divided into some number of blocks, the function should return 0.</p>
<p>For example, given:</p>
<tt style="white-space: pre-wrap;">    A[0] = 1
    A[1] = 2
    A[2] = 3
    A[3] = 4
    A[4] = 3
    A[5] = 4
    A[6] = 1
    A[7] = 2
    A[8] = 3
    A[9] = 4
    A[10] = 6
    A[11] = 2</tt>
<p>the function should return 3, as explained above.</p>
<p>Write an <b><b>efficient</b></b> algorithm for the following assumptions:</p>
<blockquote><ul style="margin: 10px; padding: 0px;"><li>N is an integer within the range [<span class="number">1</span>..<span class="number">100,000</span>];</li>
<li>each element of array A is an integer within the range [<span class="number">0</span>..<span class="number">1,000,000,000</span>].</li>
</ul>
</blockquote></div>

````python
# peak  recent_peak array   peak  
#  O(root(n)) 
#      jump  , jump index recent_peak jump  

def solution(A):
    if len(A)<3:
        return 0
        
    recent_peaks = [-1 for _ in range(len(A))]
    for idx in range(1, len(A)-1):
        if A[idx]>A[idx-1] and A[idx]>A[idx+1]:
            recent_peaks[idx] = idx
        else:
            recent_peaks[idx] = recent_peaks[idx-1]
    recent_peaks[-1] = recent_peaks[-2]

    for i in range(2, len(A)+1):
        if len(A)%i != 0:
            continue
        success=True
        for idx in range(i-1, len(A), i):
            if recent_peaks[idx]<=idx-i:
                success=False
                break
        if success:
            return len(A)//i
    return 0

solution([1])
````
