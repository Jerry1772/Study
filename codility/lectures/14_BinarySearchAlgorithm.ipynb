# Binary Search  
$O(log_2{n})$으로 정답을 찾아가는 알고리즘.  

A 안에서 x이하인 원소 중 가장 큰 원소를 찾아가는 함수
```python
def binary_search(A, x):
    n = len(A)
    beg = 0
    end = n-1
    result = -1

    while beg<=end:
        mid = (beg+end)//2
        if A[mid] <= x:
            beg = mid + 1
            result = mid
        else:
            end = mid-1
    return result
```

n개의 binary value가 있다. 1은 hole을 의미한다. k개의 크기가 같은 board가 주어질 때, 지붕의 hole들을 모두 덮을 수 있는 최소 크기를 구하라.

- size를 1부터 n 사이에서 binary search 로 구하는 방식을 사용할 수 있다.

``` python
# k size로 덮었을 때, 몇 개가 필요한지 구하는 함수
def check(A, k):
    n = len(A)
    boards = 0
    last = -1
    for i in range(n):
        if A[i]==1 and last<i:
            boards += 1
            last = i + k - 1
    return boards

# binary search 
def solution(A, k):
    n = len(A)
    beg = 1
    end = n
    while beg <= end:
        mid = (beg+end)//2
        if check(A, mid) <= k: # 충분하면 줄여본다. 정답은 갱신한다.
            end = mid-1
            result = mid
        else:
            beg = mid+1
    return result
```

# <span style="color:Salmon">MinMaxDivision</span>

<div class="brinza-task-description">
<p>You are given integers K, M and a non-empty array A consisting of N integers. Every element of the array is not greater than M.</p>
<p>You should divide this array into K blocks of consecutive elements. The size of the block is any integer between 0 and N. Every element of the array should belong to some block.</p>
<p>The sum of the block from X to Y equals A[X] + A[X + 1] + ... + A[Y]. The sum of empty block equals 0.</p>
<p>The <i>large sum</i> is the maximal sum of any block.</p>
<p>For example, you are given integers K = 3, M = 5 and array A such that:</p>
<tt style="white-space: pre-wrap;">  A[0] = 2
  A[1] = 1
  A[2] = 5
  A[3] = 1
  A[4] = 2
  A[5] = 2
  A[6] = 2</tt>
<p>The array can be divided, for example, into the following blocks:</p>
<blockquote><ul style="margin: 10px; padding: 0px;"><li>[2, 1, 5, 1, 2, 2, 2], [], [] with a large sum of 15;</li>
<li>[2], [1, 5, 1, 2], [2, 2] with a large sum of 9;</li>
<li>[2, 1, 5], [], [1, 2, 2, 2] with a large sum of 8;</li>
<li>[2, 1], [5, 1], [2, 2, 2] with a large sum of 6.</li>
</ul>
</blockquote><p>The goal is to minimize the large sum. In the above example, 6 is the minimal large sum.</p>
<p>Write a function:</p>
<blockquote><p style="font-family: monospace; font-size: 9pt; display: block; white-space: pre-wrap;"><tt>def solution(K, M, A)</tt></p></blockquote>
<p>that, given integers K, M and a non-empty array A consisting of N integers, returns the minimal large sum.</p>
<p>For example, given K = 3, M = 5 and array A such that:</p>
<tt style="white-space: pre-wrap;">  A[0] = 2
  A[1] = 1
  A[2] = 5
  A[3] = 1
  A[4] = 2
  A[5] = 2
  A[6] = 2</tt>
<p>the function should return 6, as explained above.</p>
<p>Write an <b><b>efficient</b></b> algorithm for the following assumptions:</p>
<blockquote><ul style="margin: 10px; padding: 0px;"><li>N and K are integers within the range [<span class="number">1</span>..<span class="number">100,000</span>];</li>
<li>M is an integer within the range [<span class="number">0</span>..<span class="number">10,000</span>];</li>
<li>each element of array A is an integer within the range [<span class="number">0</span>..<span class="number">M</span>].</li>
</ul>
</blockquote></div>

````python
# answer를 binary search 로 찾아가는 방식이 빠를 것 같다.
# min은 0, max는 M*len(A) 으로 시작한다. (mid = (min+max)//2)
# 매 번 A를 순회하며 mid 값 이하까지만 원소를 더한다.
# mid 값을 넘어가면 count += 1, sum = 0 을 수행한다.
# count가 k 이하면 max = mid-1, answer = mid를 수행한다.
# count가 k 초과면 min = mid+1 을 수행한다.

def check(mid, A, K):
    count = 1
    _sum = 0
    for a in A:
        if a > mid:
            return K+1
        if _sum+a > mid:
            count += 1
            _sum = a
        else:
            _sum += a
    return count
    
def solution(K, M, A):
    beg = 0
    end = M*len(A)
    answer = 0

    while beg<=end:
        mid = (end+beg)//2
        if check(mid, A, K) <= K:
            end = mid-1
            answer = mid
        else:
            beg = mid+1
    return answer

solution(3, 5, [5, 3])
````

# <span style="color:Salmon">NailingPlanks</span>

<div class="brinza-task-description">
<p>You are given two non-empty arrays A and B consisting of N integers. These arrays represent N planks. More precisely, A[K] is the start and B[K] the end of the K−th plank.</p>
<p>Next, you are given a non-empty array C consisting of M integers. This array represents M nails. More precisely, C[I] is the position where you can hammer in the I−th nail.</p>
<p>We say that a plank (A[K], B[K]) is nailed if there exists a nail C[I] such that A[K] ≤ C[I] ≤ B[K].</p>
<p>The goal is to find the minimum number of nails that must be used until all the planks are nailed. In other words, you should find a value J such that all planks will be nailed after using only the first J nails. More precisely, for every plank (A[K], B[K]) such that 0 ≤ K &lt; N, there should exist a nail C[I] such that I &lt; J and A[K] ≤ C[I] ≤ B[K].</p>
<p>For example, given arrays A, B such that:</p>
<tt style="white-space: pre-wrap;">    A[0] = 1    B[0] = 4
    A[1] = 4    B[1] = 5
    A[2] = 5    B[2] = 9
    A[3] = 8    B[3] = 10</tt>
<p>four planks are represented: [1, 4], [4, 5], [5, 9] and [8, 10].</p>
<p>Given array C such that:</p>
<tt style="white-space: pre-wrap;">    C[0] = 4
    C[1] = 6
    C[2] = 7
    C[3] = 10
    C[4] = 2</tt>
<p>if we use the following nails:</p>
<blockquote><ul style="margin: 10px; padding: 0px;"><li>0, then planks [1, 4] and [4, 5] will both be nailed.</li>
<li>0, 1, then planks [1, 4], [4, 5] and [5, 9] will be nailed.</li>
<li>0, 1, 2, then planks [1, 4], [4, 5] and [5, 9] will be nailed.</li>
<li>0, 1, 2, 3, then all the planks will be nailed.</li>
</ul>
</blockquote><p>Thus, four is the minimum number of nails that, used sequentially, allow all the planks to be nailed.</p>
<p>Write a function:</p>
<blockquote><p style="font-family: monospace; font-size: 9pt; display: block; white-space: pre-wrap;"><tt>def solution(A, B, C)</tt></p></blockquote>
<p>that, given two non-empty arrays A and B consisting of N integers and a non-empty array C consisting of M integers, returns the minimum number of nails that, used sequentially, allow all the planks to be nailed.</p>
<p>If it is not possible to nail all the planks, the function should return −1.</p>
<p>For example, given arrays A, B, C such that:</p>
<tt style="white-space: pre-wrap;">    A[0] = 1    B[0] = 4
    A[1] = 4    B[1] = 5
    A[2] = 5    B[2] = 9
    A[3] = 8    B[3] = 10

    C[0] = 4
    C[1] = 6
    C[2] = 7
    C[3] = 10
    C[4] = 2</tt>
<p>the function should return 4, as explained above.</p>
<p>Write an <b><b>efficient</b></b> algorithm for the following assumptions:</p>
<blockquote><ul style="margin: 10px; padding: 0px;"><li>N and M are integers within the range [<span class="number">1</span>..<span class="number">30,000</span>];</li>
<li>each element of arrays A, B and C is an integer within the range [<span class="number">1</span>..<span class="number"><tt style="white-space: pre-wrap;">2*M</tt></span>];</li>
<li>A[K] ≤ B[K].</li>
</ul>
</blockquote></div>

````python
# C를 binary search 대상으로 잡고
# A, B를 (a,b) 로 재구성한 뒤, 소팅한다.
# binary search 대상 C subslice 를 소팅한다.
# 이러면 C[j] 차례에는 j>i 인 C[i]가 탐색했던 (a,b)를 다시 탐색할 필요가 없으니
# 총 M+N번의 탐색으로 한 번의 check 를 수행할 수 있을 것 같다.

def check(AB, C):
    C.sort()

    # plank와 nail 각각 idx를 가지고 시작
    # plank 시작점보다 nail이 작으면, 다음 nail을 가져온다 (nail idx += 1)
    # plank 끝점보다 nail이 크면, return False 한다. (이전에 처리됐어야 함.)
    # plank 시작점과 끝점 사이에 nail이 오면, 다음 plank를 가져온다 (plank idx += 1)
    # plank idx가 끝나면 return True 한다.
    i, j = 0, 0
    while i<len(AB):
        if j==len(C):
            return False
        ab = AB[i]
        c = C[j]
        if ab[0] > c:
            j += 1
        elif ab[1] < c:
            return False
        else:
            i += 1
    return True

def solution(A, B, C):
    AB = [(a, b) for a, b in zip(A, B)]
    AB.sort()
    
    begin = 0
    end = len(C)
    answer = 0
    while begin <= end:
        mid = (begin+end)//2
        if check(AB, C[:mid]):
            end = mid-1
            answer = mid
        else:
            begin = mid+1
    if answer == 0:
        return -1
    return answer
````
