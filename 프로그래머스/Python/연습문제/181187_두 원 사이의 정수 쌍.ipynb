<p>x축과 y축으로 이루어진 2차원 직교 좌표계에 중심이 원점인 서로 다른 크기의 원이 두 개 주어집니다. 반지름을 나타내는 두 정수 <code>r1</code>, <code>r2</code>가 매개변수로 주어질 때, 두 원 사이의 공간에 x좌표와 y좌표가 모두 정수인 점의 개수를 return하도록 solution 함수를 완성해주세요.<br/>
※ 각 원 위의 점도 포함하여 셉니다.</p>

그림과 같이 정수 쌍으로 이루어진 점은 총 20개 입니다.</p>



```python
# 그냥 수학으로 하면 되지 않을까?
# -> 최대 100만개의 x값에 대해 y값의 int값 계산
# -> 걍 x값이 0부터 r포함까지 돌고
# -> 0이 여러번 빠지긴 하는데, 어차피 두 원 다 해당되니까 무시

def get_int_points(r:int, small:bool=False) -> int:
    result = 0
    for x in range(r+1):
        y = (r**2 - x**2)**0.5
        if small and y and y==int(y):
            result -= 1
        result += int(y)
    return 4*result

def solution(r1:int, r2:int) -> int:
    return get_int_points(r2)-get_int_points(r1, True)
```
