```python
a = [[1,2,3],[4,5,6],[7,8,9]]
a = [list(z) for z in zip(*a)]
print(a)
a = [list(z) for z in zip(*a)]
print(a)
```

```python
# 단순구현 문제같다.

EMPTY_FLAG = ' '

def get_coords(row, col):
    return (row, col), (row, col+1), (row+1, col), (row+1, col+1)

def get_val(board, row, col):
    return board[row][col]

def set_val(board, *coords):
    for row, col in coords:
        board[row][col] = EMPTY_FLAG

def pop(board, m, n):
    pop_set = set()
    for row in range(m-1):
        for col in range(n-1):
            a, b, c, d = get_coords(row, col)
            if get_val(board, *a)==get_val(board, *b)==get_val(board, *c)==get_val(board, *d) and\
                get_val(board, *a) != EMPTY_FLAG:
                pop_set |= set([a,b,c,d])
                
    return pop_set

def rotate(board):
    return [list(z) for z in zip(*board)]

def drop(board):
    temp_board = rotate(board)
    for b in temp_board:
        b.sort(key=lambda x: 0 if x==EMPTY_FLAG else 1)
    return rotate(temp_board)

def count(board):
    answer = 0
    for b in board:
        for bb in b:
            if bb == EMPTY_FLAG:
                answer += 1
    return answer

def solution(m, n, board):
    board = [list(b) for b in board]

    while True:
        pop_set = pop(board, m, n)
        if not pop_set:
            break
        set_val(board, *pop_set)
        
        board = drop(board)

    return count(board)

solution(6, 6, ["TTTANT", "RRFACC", "RRRFCC", "TRRRAA", "TTMMMF", "TMMTTJ"])
```

```python
# drop 을 구현하다보니까 그냥 처음부터 transpose 해서 pop과 append로 수행하면 될 것 같다.
EMPTY_FLAG = ' '
def check(board, row, col) -> bool:
    if board[row][col] == EMPTY_FLAG:
        return False
    if board[row][col]==board[row+1][col]==board[row][col+1]==board[row+1][col+1]:
        return True
    return False
def set_empty(board, row, col):
    board[row].pop(col)
    board[row].append(EMPTY_FLAG)

def solution(m, n, board):
 
    board = [list(b)[::-1] for b in zip(*board)]

    popped = True

    while popped:
        to_pop = set()
        popped = False
        for row in range(n-1):
            for col in range(m-1):
                changed = check(board, row, col)
                if changed:
                    popped = True
                    to_pop |= set([(row, col), (row+1, col), (row, col+1), (row+1, col+1)])
        for coord in sorted(to_pop, key=lambda x: x[1], reverse=True):
            set_empty(board, *coord)

    answer = 0
    for b in board:
        answer += b.count(EMPTY_FLAG)
    return answer
solution(	4, 5, ["CCBDE", "AAADE", "AAABF", "CCBBF"])
```
