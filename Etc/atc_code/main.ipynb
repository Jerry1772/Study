```python
import requests

url = 'https://www.health.kr/searchDrug/ajax/ajax_atccode.asp?select_mode={}&req_word={}'

result_list = []

def main(idx:int=1, code:str=""):
    print(f'processing {code}')
    resp = requests.get(url.format(idx, code)).json()
    result_list.extend(resp)

    for r in resp:
        if r['datatree'] and idx<5:
            main(idx+1, r['atc_code'])
```

```python
main()
```

```python
import json
with open('./result.json', 'w') as f:
    json.dump(result_list, f, ensure_ascii=False, indent=2)
```
