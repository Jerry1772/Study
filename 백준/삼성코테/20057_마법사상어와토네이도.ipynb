```python


class Board:
    _NEXT_DIR = {
        "left": "down",
        "down": "right",
        "right": "up",
        "up": "left"
    }
    _NEXT_MOVE = {
        "left": (0,-1),
        "down": (1,0),
        "right": (0,1),
        "up": (-1,0)
    }


    def __init__(self, n, map):
        self.size = n
        self.map = map

        self.row = self.size//2
        self.col = self.size//2
        self.dir = "left"

        self.out_sand = 0

        self._PROPORTION = {
            "left": [
                [0,0,0.02,0,0],
                [0,0.1,0.07,0.01,0],
                [0.05,0,0,0,0],
                [0,0.1,0.07,0.01,0],
                [0,0,0.02,0,0]
            ]
        }
        for dir in ("up", "right", "down"):
            self._PROPORTION[dir] = list(zip(*self._PROPORTION[self._NEXT_DIR[dir]][::-1]))

    @property
    def proportion(self):
        return self._PROPORTION[self.dir]
    
    @property
    def sand_amount(self):
        return self.map[self.row][self.col]
    
    def clear(self):
        self.map[self.row][self.col] = 0
        
    def _is_in_board(self, row, col):
        if 0<=row<self.size and 0<=col<self.size:
            return True
        return False

    def move(self):
        if self.dir == "left" and self.row>self.col or \
            self.dir == "down" and self.size-self.row == self.col+1 or \
            self.dir == "right" and self.row==self.col or \
            self.dir == "up" and self.row+1 == self.size-self.col:
                self.dir = self._NEXT_DIR.get(self.dir)
        
        d_r, d_c = self._NEXT_MOVE.get(self.dir)
        self.row += d_r
        self.col += d_c

    def __call__(self):
        while self._is_in_board(self.row, self.col):
            self.move()
            sand_remain = self.sand_amount
            for dr in range(5):
                for dc in range(5):
                    
                    nr, nc = self.row+dr-2, self.col+dc-2
                    sand = int(self.sand_amount*self.proportion[dr][dc])
                    sand_remain -= sand
                    if self._is_in_board(nr,nc):
                        self.map[nr][nc] += sand
                    else:
                        self.out_sand += sand
                        
            _nr, _nc = self._NEXT_MOVE[self.dir]
            nr, nc = self.row+_nr, self.col+_nc
            if self._is_in_board(nr,nc):
                self.map[nr][nc] += sand_remain
            else:
                self.out_sand += sand_remain

            self.clear()

    ## 이 방식으로 진행하려면 
```

```python
N = int(input())
sand = [list(map(int, input().split())) for _ in range(N)]
b = Board(N, sand)

b()

b.out_sand
```

```python
def recount(time, dx, dy, direction):
    global ans, s_x, s_y

    # y좌표 계산 & x좌표 갱신
    for _ in range(time):
        s_x += dx
        s_y += dy
        if s_y < 0:  # 범위 밖이면 stop
            break

        # 3. a, out_sand
        total = 0  # a 구하기 위한 변수
        for dx, dy, z in direction:
            nx = s_x + dx
            ny = s_y + dy
            if z == 0:  # a(나머지)
                new_sand = sand[s_x][s_y] - total
            else:  # 비율
                new_sand = int(sand[s_x][s_y] * z)
                total += new_sand
            if 0 <= nx < N and 0 <= ny < N:   # 인덱스 범위이면 값 갱신
                sand[nx][ny] += new_sand
            else:  # 범위 밖이면 ans 카운트
                ans += new_sand


N = int(input())
sand = [list(map(int, input().split())) for _ in range(N)]

# 2. 방향별 모래 비율 위치
left = [(1, 1, 0.01), (-1, 1, 0.01), (1, 0, 0.07), (-1, 0, 0.07), (1, -1, 0.1),
         (-1, -1, 0.1), (2, 0, 0.02), (-2, 0, 0.02), (0, -2, 0.05), (0, -1, 0)]
right = [(x, -y, z) for x, y, z in left]
down = [(-y, x, z) for x, y, z in left]
up = [(y, x, z) for x, y, z in left]

s_x, s_y = N//2, N//2  # 시작좌표(x좌표)
ans = 0  # out_sand

# 1.토네이도 회전 방향(y위치)
for i in range(1, N + 1):
    if i % 2:
        recount(i, 0, -1, left)
        recount(i, 1, 0, down)
    else:
        recount(i, 0, 1, right)
        recount(i, -1, 0, up)

print(ans)
```
